```python
import xml.etree.ElementTree as ET

def get_num_vehicles(routes_file):
    # Parse the routes.xml file
    tree = ET.parse(routes_file)
    root = tree.getroot()
    
    # Find all vehicle elements and return the count
    vehicles = root.findall('vehicle')
    return len(vehicles)

# Example usage
routes_file = 'routes.xml'
num_vehicles = get_num_vehicles(routes_file)
print(f"Number of vehicles: {num_vehicles}")
```

```python
import xml.etree.ElementTree as ET

class Vehicle:
    def __init__(self, vehicle_id):
        self.id = vehicle_id
        # You can add other attributes like route, position, etc. if needed
        self.route = []
        self.position = None
        self.speed = None

    def __repr__(self):
        return f"Vehicle(id={self.id}, route={self.route}, position={self.position}, speed={self.speed})"

def create_vehicle_list(routes_file):
    """
    Parses the routes.xml file and creates a list of Vehicle objects with their IDs.
    """
    # Parse the routes.xml file
    tree = ET.parse(routes_file)
    root = tree.getroot()

    # Create an empty list for vehicles
    vehicles = []

    # Loop through all the vehicle elements in the routes.xml file
    for vehicle in root.findall('vehicle'):
        # Extract the vehicle id
        vehicle_id = vehicle.get('id')
        
        # Create a Vehicle object and append it to the vehicles list
        new_vehicle = Vehicle(vehicle_id)
        
        # You can extract other attributes as well (route, position, speed) and add to the vehicle object
        # Example: Extracting route data if available
        route = vehicle.find('route')
        if route is not None:
            new_vehicle.route = route.attrib.get('edges', '').split()  # Assuming route is given as space-separated edges

        # Add to the list
        vehicles.append(new_vehicle)

    return vehicles

# Example usage:
routes_file = 'routes.xml'
vehicles = create_vehicle_list(routes_file)

# Print the list of vehicles
for vehicle in vehicles:
    print(vehicle)
```
